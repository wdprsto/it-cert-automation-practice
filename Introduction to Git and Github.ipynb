## Intro ke course
VCS penting untuk menyimpan histori perubahan yang terjadi dalam sistem

Keyword:
- `diff`
- `patch`

### keeping historical copies
mempertahankan histori data. git dapat ngetrack perubahan, dan bisa untuk kolaborasi.

- `diff` digunakan untuk membandingkan dua file/direktori. hanya menampilkan perbedaan dari 2 file
    `<` berarti yang dikurangi
    `>` yang ditambahkan
- diff menghasilkan masing-masing sectionyang mengalami perubahan. Misalnya pertama ada `5c5,6` itu perubahan (changed) dua baris (baris 5-6), lalu kedua ada `11a13,15` itu penambahan (added) 3 baris (baris 13-15)
- `diff -u`  untuk menampilkan perbedaan dalam format lain, `diff -u file1 file2` `@@ -2,7 +2,8 @@` itu artinya kita ngurangi baris 2-7 di file pertama, lalu nambahkan di baris 2-8 di file kedua
- `wdiff` digunakan untuk menghighlight line by line perubahan
- aplikasinya seperti ` kdiff3, meld, vimdiff`

### menggunakan diff dn patch
menggunakan perubahan untuk diterapkan
- `diff -u old_file new_file > change.diff`
diff file disini menampilkan perubahan dari file lama ke file baru, dengan tambahan konteks terhadap perubahan

-  `patch` menerapkan perubahan dari file diff ke file yang diinginkan, menggunakan standard input
`patch file_yang_mau_di_patch < diff_file`

dengan menggunakan patch, kita dapat memastikan yang diubah hanyalah yang bersesuaian. kalau kita ngubah seluruh file, bisa jadi pengirim menggunakan file template versi lama

### studi kasus diff dan patch
tahapannya
- buat diff file menggunakan `diff >`
- implementasikan dengan `patch <`

## Branching and Merging
branches penting dalam workflow git. Brach meruapkan pointer ke commit speifik, merepresentasikan independent line dari sebuah jalur pengembangan project.

Default branch yang dibuat git adalah `master`. Branch master berisi fitur terbaik dari kode kita. Untuk pengembangan, kita bisa mmebuat branch baru.

Kita dapat mengembangkan fitur baru di branch tambahan. Ketika fitur sudah benar, kita bisa `merge` kontennya di dalam `master` branch.

### Membuat branch baru
`git branch`
- untuk mengelist/menampilkan semua branch yang ada

`git branch nama_branch_baru`
- membuat branch baru berdasarkan master branch saat ini
- tampilkan lagi apa branch yang digunakan saat ini dengan `git branch`

`git checkout`
- sebelumnya, kita gunakan ini untuk kembali dari non-stage file kembali ke asal.
- kita gunakan `git checkout` untuk nge-check out snapshot terakhir dari files maupun branches

`git checkout nama_branch`
- berpindah ke branch tujuan

`git checkout -b nama_branch`
- untuk membuat branch baru dan langsung berpindah ke branch tersebut

kita bisa menambahkan file baru di branch tersebut. Kita coba membuat file baru, lalu commit pada branch baru.
cek isinya dengan `git log-2`, akan ditampilkan bahwa HEAD saat ini mengacu pada brach baru, dan commit berhasil dilakukan terhadap branch baru tersbut.

### Bekerja denagn branch
`git checkout master`
- kembali ke branch utama. semua file yang ada didalam direktori kita akan menyesuaikan dengan isi dari branch tersebut.

`git branch -d nama_branch`
- menghapus sebuah branch yang ada.
- saat ada data yang belum di merge, namun kita mencoba untuk menghapusnya, kita akan diinformasikan langkah selanjutnya, yaitu menggunakan flag `-D` jika benar" ingin menghapus branch

### Merging Branch
merging => istilah git yang digunakan untuk mengkombinasikan/menggabungkan branched data & history bersama-sama.

`git merge`
- menggabungkan snapshot & history dari satu branch ke branch master

tahapan merge:
- pindah ke master branch `git branch master`
- gabungkan dengan `git merge branch_yang_ingin_digabungkan`
- selanjutnya, proses merge pun akan dilakukan
- lihat statusnya dengan `git log -p`, tampak bahwa kedua branch sekarang merujuk pada pointer/HEAD yang sama

Git menggunakan dua algoritma untuk merge: `fast-forward dan  three way merge`.
- `fast forward` => ketika semua commit sama, tidak diverge (menyebar).
- `three way` => ketika merging branch sudah diverge, memiliki commit lainnya sebelum dimerge. model ini terjadi ketika terjadi komit di salah satu branch setelah kedua branch tersebut dipisah (split)

Jika terdapat perbedaan antara 2 snapshot pada file berbeda/lokasi berbeda, maka bisa di merge (kedua perubahan akan digabungkan). Jika perubahan terjadi di file yang sama, di posisi yang sama, maka git tidak dapat menangani konflik ini secara otomatis. Kita perlu menanganinya secara manual. (`conflict`)

### Merge Conflict
Untuk mengatasi konflik, kita bisa gunakan command `git status`.

tahapan
- buka file yang berkonflik. git akan menambahkan informasi terkait konflik yang ada. kita pilih mana yang mau dipertahankan/ dihapus.
- pertahankan baris yang ingin digunakan. hapus juga simbol penanda branchnya (<<<<<< ====== >>>>>>). Setelah sudah, simpan
- jalankan `git add nama_file_yang_konflik`
- selesaikan dengan `git commit`

`git log --graph --oneline`
- untuk melihat commit dalam bentuk graph gunakan flag `--graph`, dan gunakan `--oneline` untuk melihat hanya satu line per commit

`git merge --abort`
- untuk throw merge dan mulai dari awal, kita gunakan kode di atas. ini akan menghentikan proses merge dan mengembalikan kondisi file seperti sebelum proses merge

## Solving Conflict
### Pull Merge Push workflow
conflict terjadi jika kita mengubah file di repo local dan remote, pada file dan posisi yang sama.

- kita perlu mengintegrasikan remote repository sebelum kita push dengan `git pull`. Nanti, akan muncul `conflict`. Untuk melihatnya, kita coba lihat visualisasi commit dengan `git log --graph --oneline --all`. Untuk menangani konflik, kita perlu menggunakan `three-way merge`

mengatasi:
- `git pull` lebih dulu sebelum nge push. jika terjadi error, lanjut ke step 2

- `git log -p origin/main`
kita lihat perubahan yang terjadi pada commit tersebut dengan `git log -p origin/main`. akan muncul baris yang diubah di remote repo, yang mungkin juga mau kita ubah sgh mengakibatkan konflik

- perbaiki konflik secara manual dengan mempertahankan salah satu kode yang benar, diantara 2 yang sedang konflik

- karena tadi konflik, maka untuk merge nya kita lakukan `git add file_konflik`, lalu `git commit`

- terakhir, push ke remote server lagi `git push`

- tampilkan hasil perubahan dengan `git log --graph --oneline`

### Rebase Your Change: MERGING FEATRURE BRANCHES TO MAIN TRUNK + MAKE IT LINEAR
kita dapat menyatukan kembali branch cab ang dengan master branch dengan menggunakan
- `git merge`
- `git rebase nama_new_base`: mengganti base commit yang kita gunakan untuk branch.
git rebase bekerja dengan mengubah asal spil branch ini dari indukan ke anakannya yang konflik tadi. kita akhirnya bisa melakukan kommit di atas base yang baru.
memunggkinkan untuk `fast forward merge`.

`git rebase`
- kita mencoba untuk me-replay kommit kita setelah komit terakhir di branch. ini akan bekerja secara otomatis jika perubahan terjadi pada bagian file yang berbeda. namun, akan memerlukan intervensi jika perubahan terjadi di file yang berbeda.

contoh: rebase refactoring branch ke main branch.
- kita lakukan commit pada file A di branch main. lalu, kita juga lakukan commit file A di branch refactoring, pada file yang sama (lokasi berbeda)  . Akibatnya, tentu akan ada  conflict saat merge. Kita bisa gunakan rebase
- pindah ke branch anakan dengan `git checkout refactoring`
- `git rebase master`, kita melakukan rebase branch 'refactoring' saat ini terhadap branch anakan master. output dari branch rebase ini adalah bahwa master branch menjadi LINEAR dengan perubahan di branch 'refactoring'
- setelah master branch linear, kita bisa merge commit di branch main dengan `git checkout main`
- kemuian, eksekusi `git merge refactoring` untuk me-merge keuda branch tadi yang udah di rebase

Dengan bantuan `git rebase`, kita dapat membuat history kita tetap linear dengan fast-forward merge.

`git push --delete origin nama_branch_mau_dihapus`
- menghapus sebuah branch pada remote repo
`git branch -d nama_branch_mau_dihapus`
- menghapus branch pada local repository

setelah branch refactoring dihapus, kita bisa melakukan `git push` terhadap remote repo kita.

### Contoh lain rebase: FETCH - REBASE - PUSH (pada branch yang sama)
kita ingin merebase ke master change ketika orang lain juga melakukan perubahan. kita ingin tetap linear, jadi jangan gunakan `git pull`, karena ia membuat commit baru. Ingin menyimpan history dengan linear

- modifikasi file, lalu commit.
- gunakan `git fetch` yang akan mengambil perubahan terakhir dari file, tidak langsung git pull.
- `git rebase origin main`.  untuk membuat commit linear pada repositori ini, setelah di commit dengan data dari remote repo. ketika ada konflik,  maka diberikan pemberitahuan conflict. Kita perbaiki dengan manual file yang konflik ini.
- untuk menyelesaikan rebase, lanjutkan dengan `git rebase --continue`
- saat dicek dengan `git log --graph --oneline`, history akan menjadi linear
- push history ke remote repo dengan `git push`

`git rebase` juga bisa digunakan untuk nge-squash 2 commit jadi 1, serta mengubah urutan commit.

## Pull Request
### Simple Pull Request
edit file secara langsung:
- `fork` file tersebut, yaitu membuat copy terhadap sebuah repository, sehingga kita bisa melakukan pull terhadap projek yang bukan milik kita.

'Pull request'
- commit/ series of commit yang kita kirimkan ke pemilik repository sehingga mereka bisa meng-inkorporasikan repo ini ke tree mereka.

Tutorial
- kita fork sebuah repo
- modifikasi file pada repo hasil fork
- mempropose perubahan file dengan klik "Propose file change". kita sudah meng-commit di forked repo kita, selanjutnya kirim `pull request` yang akan mengirim perubahan ke repo asal.
- untuk mengirim perubahan ke tree utama, klik `create pull request`. kita bisa memberikan komentar dari pull request kita. Checkbox `allow edits from maintainers`, sehingga si maintainers bisa mengedit kode dari kita jika ada penyesuaian.
- kita membuat pull request, sehingga owner bisa memilih untuk merge atau tidak.

### Typical Pull Request

Tahapan:
- melakukan fork terhadap repo dengan menekan tombol `fork`
- salin repo di local repo dengan menggunakan command `git clone`
- direksikan terminal ke folder dengan `cd rearrange`
- lihat history commit dengan `git log`
- lakukan modifikasi sebagai berikut
    = buat branch baru dengan nama add-readme `git checkout -b add-readme`
    = tambah file README.md
    = commit perubahan ini dengan `git add` lalu `git commit`
- agar dapat menggabungkannya ke main tree, kita harus push dulu branch baru ini ke remote repo kita. commandnya
`git push -u origin nama_branch_baru`
- lakukan `pull request` dengan
    = klik `pull requests` pada github
    = pastikan ada keterangan `able to merge`. jika tidak bisa, maka harus di `rebase` dengan current branch original repo. kita bisa melihat diff yang ada di bagian bawah halaman
    = sudah final, klik `create pull request`


### Update current pull request
Untuk mengubah pull request yang sudah kita buat, kita perlu mengubah dari local repo dulu
- edit perubahan pada local repository
- lakukan commit `git commit -a -m "Add more information xxx"
- push perubahan ke remote repository dengan `git push`

Dengan commit di branch local yang sama, maka github secara otomatis menambahkan perubahan ke pull request yang sebelumnya.

### Squash perubahan jadi single commits
kita sebaiknya tidak me-rewrite history ketika commit sudah di publish. kita mau membuat single commit yang meliputi 2 perubahan

`rebase -i master_branch`
- memilih semua commit dari yang paling lama ke yang paling baru, kemudian menerapkan operasi tertentu terhadap masing-masing commit yang tertera di layar. yang di rebase untuk squash ialah master brancnya, yaitu `main`
- akan ditampilkan beberapa opsi, seperti `pick` untuk menggunakan commit, `reword` untuk menggunakan commit dengan message yang diubah, dll.
jadi, yang berpengaruh adalah `apa isi commit messagenya`
- jika kita memilih `squash`, commit message akan digabung.
- jika memilih `fixup`, commit message akan dihapus

contoh implementasi
- 
```
//awal, defaultnya pick
pick 736d754
pick 01231b0
//perubahan
pick 736d754
squash 01231b0
//tuliS di satu terakhir saja, karena itu artinya dia menggabungkan yang ini dengan yang sebelumnya
```

saat kita memilih meng-squash commit ke satu sebelumnya, kita masuk ke text editor.
kita bisa menyesuaikan commit messagenya

- simpan perubahan. lihat riwayat perubahan dengan `git show`
- cek info saat ini dengan `git status`. lihat graph dengan `git log --graph --oneline --all -4` untuk melihat 4 commit terakhir
- kita ingin me-replace old commit di remote repo dengan commit gabungan yang baru kita buat, caranya
`git push -f` (force git to push it)


## Code Reviews
### Apa itu code review?
berfungsi untuk mereview file. Melakukan `code review` artinya memahami kode, dokumentasi, atau konfigurasi orang lain, dan mengecek apakah semuanya masuk akal dan mengikuti pattern yang diharapkan.

Review meningkatkan kualitas kode karena akan banyak orang yang mereview kodenya.

Masalah code review umum
- unclear name yang membuat orang lain sulit memahami
- forgetting to add a test
- forgetting to add a specific condition

Github
- code review hanya ditujukan untuk orang yang tidak punya akses (via pull request)

### Code review workflow


- `nit` memahami apa makna dari comment yang diberikan reviewer, dan lakukan perbaikan yang sesuai dengan komen tersebut. Ambil ini sebagai opportunity untuk membuat kode lebih baik.
In git jargon (and elsewhere in the tech world), a nit is a minor “nitpick” about a piece of code. Nit dapat diartikan sebagai A trivial comment or suggestion
- gunakan style guide untuk python, misalnya PEP8.
- ada tool yang memerlukan accept dari reviewer agar bisa usbmit, tapi ada yang hanya perlu +1.

### How to use code review in github

- code review memiliki komentar baris per baris.

tahapan:
- setelah mendapatkan komentar/review di github, perbaiki file di local.
- karena kita ingin commit ini menjadi bagian dari commit sebelumnya, maka gunakan command
`git commit -a --amend`
- cek status repo, gunakan `git status`
- karena kita melakukan amend, maka untuk mempushnya ke remote repository gunakan
`git push -f`

nanti, status komentarnya menjadi 'outdated', dan kita bisa klik tombol `resolved conversation`, dan memberikan komentar kalau kita sudah menyelsaikan semua komentar yang diberikan.

### Continuous Integration
`CI` menjalankan kode tes sistem secara otomatis. CI akan membangun dan mengetest kode kita setiap kali ada perubahan (tiap ada commit di branch repository kita)

Saat kode kita dibangun dan di tes, tahap otomasi sealnjutnya adalah `continous deployment`, yang disebut sebagai CD (continous delivery/deployment)
goal:
- tiap kode sering di upload/diperbarui
- melakukan inkremental update dengan hanya sedikit  perubahan pada waktu itu

`CI/CD`
- tools yang bergantung dengan ci/cd. CD akan melakukan update, dan dipaatkan
- psinya adalah JenKins.

`Travis` mampu mengenali integrasi mana yang harus bekerja
